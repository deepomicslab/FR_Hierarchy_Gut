Compare nFR/SE between EN and EB groups with wilcox testing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nfr = pd.read_csv('../result/Anti/nFR/nfr_df.tsv', sep='\t', index_col=0, header=0)
se = pd.read_csv('../result/Anti/SE/se_df.tsv', sep='\t', index_col=0, header=0)
p_cutoff = 0.05
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu

def compare_groups(df, title, col, g1, g2):
    
    anti = pd.read_csv("../data/Anti/Anti.group.tsv", sep="\t")
    
    sample_g1 = anti.loc[anti[col] == g1]["sample_id"]
    sample_g2 = anti.loc[anti[col] == g2]["sample_id"]
    
    p_values = []
    cluster_names = []
    enriched_groups = []
    mean_g1 = []
    mean_g2 = []
    for col in df.columns:
        values_g1 = df[col][sample_g1]
        values_g2 = df[col][sample_g2]
        _, p_value = mannwhitneyu(values_g1, values_g2)
        p_values.append(p_value)
        cluster_names.append(col)
        
        if values_g1.mean() > values_g2.mean():
            enriched_groups.append(g1)
        else:
            enriched_groups.append(g2)
        
        mean_g1.append(values_g1.mean())
        mean_g2.append(values_g2.mean())
    
    result = pd.DataFrame({"FR": [title] * len(cluster_names), 
                           "Group1": [g1] * len(cluster_names), 
                           "Group2": [g2] * len(cluster_names), 
                           "Cluster": cluster_names, 
                           "p_value": p_values,
                           "enriched": enriched_groups,
                           "mean_g1": mean_g1,
                           "mean_g2": mean_g2})
    return result
```

```python
testing_results = pd.DataFrame()


testing_results = pd.concat([testing_results, compare_groups(nfr, "nFR", "ECC_day", "EB_0", "EN_0")])
testing_results = pd.concat([testing_results, compare_groups(nfr, "nFR","ECC_day", "EB_7", "EN_7")])
testing_results = pd.concat([testing_results, compare_groups(nfr, "nFR","ECC_day", "EB_90", "EN_90")])
significant_results = testing_results[testing_results["p_value"] < p_cutoff]

significant_results.to_csv("../result/Anti/nFR/nfr.EB_EN.differential.tsv",sep="\t",index=False)
significant_results
```

```python
testing_results = pd.DataFrame()


testing_results = pd.concat([testing_results, compare_groups(se, "SE", "ECC_day", "EB_0", "EN_0")])
testing_results = pd.concat([testing_results, compare_groups(se, "SE","ECC_day", "EB_7", "EN_7")])
testing_results = pd.concat([testing_results, compare_groups(se, "SE","ECC_day", "EB_90", "EN_90")])
significant_results = testing_results[testing_results["p_value"] < p_cutoff]

significant_results.to_csv("../result/Anti/SE/SE.EB_EN.differential.tsv",sep="\t",index=False)
significant_results
```
